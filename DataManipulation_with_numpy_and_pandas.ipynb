```python
## Handling missing values
df.isnull()   ## wherever value exists we get False. but wherever the value is missing or novalue we get True
```

```python
df.fillna(0) ## replaces all missing values with value 0
```

```python
## filling missing values with mean of column

df['Sales_fillNA'] = df['Sales'].fillna(df['Sales'].mean())
df
```

```python
df.dtypes
```

```python
## Renaming Columns
df = df.rename(columns={'Date':'Sales Date'})
df.head(5)
```

```python
## change datatype
df['Value_new'] = df['Value'].astype(int)  ## IntCastingNaNError: Cannot convert non-finite values (NA or inf) to integer. since values are not available unable to convert. hence first we replace the values with anyother vaue
df.head()   
```

```python
df['Value_new'] = df['Value'].fillna(df['Value'].mean()).astype(int)  # we have created a new column and filled data with its mean and converted to another datatype
df.head()
```

```python
df['New Value'] = df['Value_new'].apply(lambda x:x*2)
df.head()
```

```python
## Data Aggregating and Grouping

df.head()
```

```python
grouped_mean = df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum = df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

```python
## aggregate muliple functions

grouped_agg = df.groupby('Region')['Value'].agg(['mean','sum','count'])
print(grouped_agg)
```

```python
### Merging and Joining DataFrames

df1 = pd.DataFrame(
    {
        'Key':['A','B','C'],
        'Value1':[1,2,3]
    }
)
df2 = pd.DataFrame(
    {
        'Key':['A','B','D'],
        'Value2':[4,5,6]
    }
)
df1
```

```python
df2
```

```python
## Merge DataFrame on the 'Key Columns'

pd.merge(df1,df2,on="Key", how="inner")
```

```python
pd.merge(df1,df2,on="Key", how="outer")
```

```python
pd.merge(df1,df2,on="Key", how="left")
```

```python
pd.merge(df1,df2,on="Key", how="right")
```
